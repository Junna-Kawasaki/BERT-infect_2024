## Count sequence number in the reference alignment

```python
def count_signs(filename):
    count = 0
    with open(filename, 'r') as file:
        for line in file:
            if '>' in line:
                count += 1
    return count
```

```python
tree_fasta_f = "/".join(map(str, ["input_examples", "Coronaviridae", "ictv.fasta"]))
print (count_signs(tree_fasta_f))
```

## convert nucleotide sequences to protein sequences using tBLASTn

```python
%%bash

train_virus="Coronaviridae"
pred_virus="sarbecov"

mkdir output_examples/${virus}
        
## リファレンス配列からギャップを除く
cat input_examples/${virus}/ictv.fasta | \
    seqkit seq -g \
    > output_examples/${virus}/ictv.ref.fasta

## 配列ファイルの指定
fasta_f=../002_Data_splitting/input_examples/${virus}.curated.fasta
        
## Accession Noの取り出し
cat input_examples/final_dataset_phylogenetic_tree_label.txt | \
    grep -e ${train_virus} -e ${pred_virus} | awk 'BEGIN{FS="\t"}NR>2{print $1}' \
    > output_examples/${virus}/accession.txt

## 配列取り出し
cat ${fasta_f} | \
    seqkit seq -g | \
    seqkit grep -f output_examples/${virus}/accession.txt \
    > output_examples/${virus}/add.fasta

## blast DB作成
makeblastdb \
    -in output_examples/${virus}/add.fasta \
    -out output_examples/${virus}/add.DB \
    -dbtype nucl

## blastp
tblastn \
    -query output_examples/${virus}/ictv.ref.fasta \
    -db output_examples/${virus}/add.DB \
    -out  output_examples/${virus}/tblastn.txt \
    -outfmt '6 qseqid sseqid evalue bitscore length qlen slen pident sseq' 
```

```python
virus = virus_l[0]

## read blast result
blast_l = []

blast_f = "/".join(map(str, ["output_examples", virus , "tblastn.txt"]))
df_blast = pd.read_csv(blast_f, sep="\t", 
                       names=["query", "subject", "evalue", "bitscore", "alen", "qlen", "slen", "pident", "sseq"])
    
df_blast = df_blast.astype({"alen":int, "qlen": int, "bitscore":float})
df_blast["coverage"] = df_blast["alen"] / df_blast["qlen"] * 100
```

```python
# select tophit sequence
idx = df_blast.groupby('subject')['bitscore'].idxmax()
result = df_blast.loc[idx]
```

```python
## output protein sequence

# read virus_ID
df_tree = pd.read_csv("input_examples/final_dataset_phylogenetic_tree_label.txt",
                      sep="\t")
output_l = df_tree["virus_ID"].unique()
df_subject = result[result["subject"].isin(output_l)]
df_subject = df_subject.loc[:, ["subject", "sseq"]]

# write protein sequences
output_f =  "/".join(map(str, ["output_examples", virus_l[0], "tblastn.fasta"]))
print (output_f)

f = open(output_f, "w")
for i, row in df_subject.iterrows():
    header, seq = row[:2]
    f.write(">" + header + "\n")
    f.write(seq + "\n")

f.close()
```

# mafft alignment

```python
%%bash

train_virus="Coronaviridae"
        
## remove gaps from reference alignment
cat input_examples/${train_virus}/ictv.fasta | \
    seqkit seq -g \
    > output_examples/${train_virus}/ictv.ref.fasta
        
## mafft
mafft --auto \
    output_examples/${train_virus}/ictv.ref.fasta \
    > output_examples/${train_virus}/ictv.ref.mafft.fasta
        
## mafft -add
mafft --auto --keeplength \
    --addfragments output_examples/${train_virus}/tblastn.fasta \
    --thread 4 \
    output_examples/${train_virus}/ictv.ref.mafft.fasta \
    > output_examples/${train_virus}/tblastn.mafft.fasta
        
## extract reference sequence names
cat output_examples/${train_virus}/ictv.ref.fasta | \
    grep ">" | sed 's/>//g' \
    > output_examples/${train_virus}/ictv.ref.txt
        
## remove reference sequences
cat output_examples/${train_virus}/tblastn.mafft.fasta | \
    seqkit grep -v -f output_examples/${train_virus}/ictv.ref.txt \
    > output_examples/${train_virus}/tblastn.mafft.rmdup.fasta
```

# phyogenetic analysis using iqtree

```python
%%bash

train_virus="Coronaviridae"

iqtree \
    -s output_examples/${train_virus}/tblastn.mafft.rmdup.fasta \
    -m TESTNEW \
    -redo \
    -nt AUTO
```
